### Visualization of the stress:

```python
#Define a random force vector (dF)
dF = np.array([0.1, 0.3, 0.5])
dF = dF/np.linalg.norm(dF)
print(f'Force vector dF: {dF}')

#Random plane normal (dS) and a random force (dF)
dS = np.array([0.1, 0.5, 0.2])
n = dS/np.linalg.norm(dS)
print(f'Plane normal vector n: {n}')

#Plane center point
O = np.array([0.5, 0.5, 0.5])
print(f'Plane center point O: {O}')
```

```python
#Defining the plane basis (u,v)

#I will choose a non-parallel vector to dS to create the plane basis (u,v)
a = np.array([0.5, 0.1, 0.2])
a = a/np.linalg.norm(a)

print(f'If the vectors are parallel a*n = 1:')
print(f'a*n = {np.dot(n, a)}\n')

#We project a onto the plane
u = a - np.dot(a, n)*n
u = u/np.linalg.norm(u)
print(f'First basis vector u: {u}')

#The second basis vector v is obtained by the cross product
v = np.cross(n, u)
v = v/np.linalg.norm(v)
print(f'Second basis vector v: {v}')
```

```python
#Plane meshgrid coordinates
s = np.linspace(-1.0, 1.0, 50)
t = np.linspace(-1.0, 1.0, 50)
S, T = np.meshgrid(s, t)

X = O[0] + u[0]*S + v[0]*T
Y = O[1] + u[1]*S + v[1]*T
Z = O[2] + u[2]*S + v[2]*T
```

```python
#Plot the plane, normal (dS) and force (dF)
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

#Plane surface
ax.plot_surface(X, Y, Z, color='grey', alpha=0.5, rstride=1, cstride=1, linewidth=0)

#Point O
ax.scatter(O[0], O[1], O[2], color=colors[7], s=50)

#Plot normal (dS) and force (dF) as arrows starting at O
ax.quiver(O[0], O[1], O[2], n[0], n[1], n[2],
          color=colors[0], linewidth=2, normalize=True, arrow_length_ratio=0.1)

#Force arrow
ax.quiver(O[0], O[1], O[2], dF[0], dF[1], dF[2],
          color=colors[2], linewidth=2, normalize=True, arrow_length_ratio=0.1)

#Annotate
offset_O = 0.1
scale_ds = 1.7
scale_df = 0.9
ax.text(O[0]-offset_O, O[1], O[2]+offset_O, r'$O$',
        color=colors[7], fontsize=20, fontdict={'weight':'bold'})
ax.text(O[0]+dS[0]*scale_ds, O[1]+dS[1]*scale_ds, O[2]+dS[2]*(scale_ds+0.4), r'$d\vec{S}$',
        color=colors[0], fontsize=20, fontdict={'weight':'bold'})
ax.text(O[0]+dF[0]*scale_df, O[1]+dF[1]*scale_df, O[2]+dF[2]*(scale_df+0.15), r'$d\vec{F}$',
        color=colors[2], fontsize=20, fontdict={'weight':'bold'})

#Axis as vectors
x0, y0, z0 = 0, 0, 0
ax.quiver(x0, y0, z0, 2, 0, 0, color='black', arrow_length_ratio=0.05)  # x-axis vector
ax.quiver(x0, y0, z0, 0, 1.4, 0, color='black', arrow_length_ratio=0.05)  # y-axis vector
ax.quiver(x0, y0, z0, 0, 0, 2, color='black', arrow_length_ratio=0.05)  # z-axis vector
ax.text(2.05, -0.05, 0, r'$x$', color='black', fontsize=20, fontdict={'weight':'bold'})
ax.text(0, 1.45, 0, r'$y$', color='black', fontsize=20, fontdict={'weight':'bold'})
ax.text(0, 0, 2.05, r'$z$', color='black', fontsize=20, fontdict={'weight':'bold'})

#Limits
scale_axis = 1.25
ax.set_xlim3d(0, scale_axis*np.max(X))
ax.set_ylim3d(0, scale_axis*np.max(Y))
ax.set_zlim3d(0, scale_axis*np.max(Z))


#View and grid
ax.set_axis_off()
ax.view_init(elev=25, azim=20)

fig.subplots_adjust(left=0, right=0.01, bottom=0, top=0.1)
fig.tight_layout()
#fig.savefig(path + 'stress_explanation.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300);
```

### Components of the stress tensor:

```python
#Plot the plane, normal (dS) and force (dF)
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

#Axis as vectors
x0, y0, z0 = 0, 0, 0
ax.quiver(x0, y0, z0, 2, 0, 0, color='black', arrow_length_ratio=0.05)  # x-axis vector
ax.quiver(x0, y0, z0, 0, 2, 0, color='black', arrow_length_ratio=0.05)  # y-axis vector
ax.quiver(x0, y0, z0, 0, 0, 2, color='black', arrow_length_ratio=0.05)  # z-axis vector
ax.text(2.2, 0, 0, r'$x$', color='black', fontsize=20, fontdict={'weight':'bold'})
ax.text(0, 2.05, 0, r'$y$', color='black', fontsize=20, fontdict={'weight':'bold'})
ax.text(0, 0, 2.05, r'$z$', color='black', fontsize=20, fontdict={'weight':'bold'})

#Grids for faces
x_min, y_min, z_min, x_max, y_max, z_max = 0, 0, 0, 1, 1, 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 10), np.linspace(y_min, y_max, 10))
zz, yy2 = np.meshgrid(np.linspace(z_min, z_max, 10), np.linspace(y_min, y_max, 10))
xx2, zz2 = np.meshgrid(np.linspace(x_min, x_max, 10), np.linspace(z_min, z_max, 10))

#Faces of the box
ax.plot_surface(xx, yy, z_min*np.ones_like(xx), alpha=0.3, color='grey') #Bottom (z = z_min)
ax.plot_surface(xx, yy, z_max*np.ones_like(xx), alpha=0.3, color='grey') #Top (z = z_max)

ax.plot_surface(x_min*np.ones_like(zz), yy2, zz, alpha=0.2, color='grey') #Left (x = x_min)
ax.plot_surface(x_max*np.ones_like(zz), yy2, zz, alpha=0.2, color='grey') # Right (x = x_max)

ax.plot_surface(xx2, y_min*np.ones_like(xx2), zz2, alpha=0.4, color='grey')	#Front (y = y_min)
ax.plot_surface(xx2, y_max*np.ones_like(xx2), zz2, alpha=0.4, color='grey')	#Back (y = y_max)

#Stress arrows
c = 0.5
ax.quiver(1, c, c, 1.2, 0, 0, color=colors[6], linewidth=2.5, normalize=True, arrow_length_ratio=0.15)
ax.quiver(1, c, c, 0, 1.2, 0, color=colors[6], linewidth=2.5, normalize=True, arrow_length_ratio=0.15)
ax.quiver(1, c, c, 0, 0, 1.2, color=colors[6], linewidth=2.5, normalize=True, arrow_length_ratio=0.15)

ax.quiver(c, 1, c, 1.2, 0, 0, color=colors[0], linewidth=2.5, normalize=True, arrow_length_ratio=0.15)
ax.quiver(c, 1, c, 0, 1.2, 0, color=colors[0], linewidth=2.5, normalize=True, arrow_length_ratio=0.15)
ax.quiver(c, 1, c, 0, 0, 1.2, color=colors[0], linewidth=2.5, normalize=True, arrow_length_ratio=0.15)

ax.quiver(c, c, 1, 1.2, 0, 0, color=colors[2], linewidth=2.5, normalize=True, arrow_length_ratio=0.15)
ax.quiver(c, c, 1, 0, 1.2, 0, color=colors[2], linewidth=2.5, normalize=True, arrow_length_ratio=0.15)
ax.quiver(c, c, 1, 0, 0, 1.2, color=colors[2], linewidth=2.5, normalize=True, arrow_length_ratio=0.15)

#Annotate
offset_x = 0.22
offset_y = 0.22
ax.text(2+offset_x, c, c, r'$\sigma_{xx}$',
        color=colors[6], fontsize=20, fontdict={'weight':'bold'})
ax.text(1, c-offset_y, 1.5, r'$\sigma_{zx}$',
        color=colors[6], fontsize=20, fontdict={'weight':'bold'})
ax.text(1, 1.5, c, r'$\sigma_{yx}$',
        color=colors[6], fontsize=20, fontdict={'weight':'bold'})

ax.text(1.5+offset_x, c, 1, r'$\sigma_{xz}$',
        color=colors[2], fontsize=20, fontdict={'weight':'bold'})
ax.text(c, c, 2, r'$\sigma_{zz}$',
        color=colors[2], fontsize=20, fontdict={'weight':'bold'})
ax.text(c, 1.5, 1, r'$\sigma_{yz}$',
        color=colors[2], fontsize=20, fontdict={'weight':'bold'})

ax.text(1.5+offset_x, 1, c, r'$\sigma_{xy}$',
        color=colors[0], fontsize=20, fontdict={'weight':'bold'})
ax.text(c, 1, 1.5, r'$\sigma_{zy}$',
        color=colors[0], fontsize=20, fontdict={'weight':'bold'})
ax.text(c, 2, c, r'$\sigma_{yy}$',
        color=colors[0], fontsize=20, fontdict={'weight':'bold'})


#Limits
scale_axis = 1.95
ax.set_xlim3d(0, scale_axis)
ax.set_ylim3d(0, scale_axis)
ax.set_zlim3d(0, scale_axis)
#ax.set_box_aspect([1, 1, 1])


#View and grid
ax.set_axis_off()
ax.view_init(elev=25, azim=30)

fig.subplots_adjust(left=0, right=1, bottom=0, top=1)

#set_axes_equal(ax)
#fig.savefig(path + 'stress_components.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300);
```
